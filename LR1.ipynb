```python
import numpy as np
import matplotlib.pyplot as plt

data_all = np.genfromtxt("data.data", delimiter=",", dtype=str)

oil = []
antifrieze = []
atf = []
gasoline = []

for row in data_all:
    if len(row) == 3:
        try:
            oil.append(row[0])
            antifrieze.append(float(row[0]))
            atf.append(float(row[1]))
            gasoline.append(float(row[2]))
        except ValueError:
            print(f"Skipping invalid data row: {row}")
    else:
        print(f"Skipping invalid data row (incorrect number of columns): {row}")


for i in range(len(oil)):
    print(f"{i}\t==>\t {oil[i]} \t\t {antifrieze[i]} \t\t {atf[i]} \t\t {gasoline[i]}")


plt.figure(1)
plt.plot(antifrieze[:30], atf[:30], 'ro', label="drive")
plt.plot(antifrieze[30:60], atf[30:60], 'g^', label="going to service")
plt.plot(antifrieze[60:90], atf[60:90], 'bs', label="car is dead")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.xlabel('Oil')
plt.ylabel('Antifrieze')


plt.figure(2)
plt.plot(antifrieze[:30], gasoline[:30], 'ro', label="drive")
plt.plot(antifrieze[30:60], gasoline[:30], 'g^', label="going to service")
plt.plot(antifrieze[60:90], gasoline[:30], 'bs', label="car is dead")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.xlabel('Oil')
plt.ylabel('atf')

plt.figure(3)
plt.plot(atf[:30], gasoline[:30], 'ro', label="drive")
plt.plot(atf[30:60], gasoline[:30], 'g^', label="going to service")
plt.plot(atf[60:90], gasoline[:30], 'bs', label="car is dead")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.xlabel('atf')
plt.ylabel('Gasoline')


plt.show()
```
